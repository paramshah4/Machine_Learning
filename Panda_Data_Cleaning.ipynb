```python
#Q1
df['FlightNumber'].fillna(-1,inplace=True)
for i in range(len(df['FlightNumber'])):
    if(df['FlightNumber'].iloc[i]==-1):
        df['FlightNumber'].iloc[i]=df['FlightNumber'].iloc[i-1]+10
df['FlightNumber']=df['FlightNumber'].astype(int)
df.dtypes
```

```python
#q2
df['from']=df['From_To'].apply(lambda x: x.split('_')[0])
df['to']=df['From_To'].apply(lambda x: x.split('_')[1])
```

```python
def func(x):
    d_new = {}
    for i in range(len(x)):
        d_new['delays_'+str(i)] = x[i]
    return pd.Series(d_new)
```
